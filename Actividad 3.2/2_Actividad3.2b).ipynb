```python
#Importamos los programas necesarios
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#Exportamos el excel que vayamos a usar
mallorca=pd.read_excel("Mallorca.xlsx")
mallorca
```

```python
#Contamos cuántos valores nulos hay en el excel
valores_nulos=mallorca.isnull().sum().sum()
valores_nulos
```

```python
#Definimos una función para sustituir los valores nulos cualitativos con el método "ffill"
def sustitución_ffill(mallorca):
    sin_cambios = mallorca.select_dtypes(include=["float","int"])
    ffill_con_nulos = mallorca.select_dtypes(include=["object","datetime","category"])
    ffill_sin_nulos = ffill_con_nulos.fillna(method="ffill")
    Datos_sin_nulos = pd.concat([sin_cambios,ffill_sin_nulos],axis=1)
    return(Datos_sin_nulos)
```

```python
#Probamos que la función trabaje correctamente
sustitución_ffill(mallorca)
```

```python
#Nos aseguramos que haya un cambio real en la cantidad de valores nulos
print(valores_nulos)
valores_nulos2=sustitución_ffill(mallorca).isnull().sum().sum()
valores_nulos2
```

```python
#Creamos un nuevo excel basado en la nueva función
sustitución_ffill(mallorca).to_csv("ffill_sin_nulos.csv")
```
